```python
import csv
import re

# Function to update a single row
def update_row(row):
    # Split the row into individual elements
    elements = row.split(',')
    
    # Initialize variables to store summed values
    sum_2nd_column = 0
    sum_3rd_column = 0
    sum_4th_column = 0
    sum_5th_column = 0
    
    # Iterate through the elements
    for i, element in enumerate(elements):
        # Check if the element starts with "ingr"
        if re.match(r'^ingr', element):
            # If so, add its corresponding value to the respective sum
            sum_2nd_column += float(elements[i + 3])
            sum_3rd_column += float(elements[i + 4])
            sum_4th_column += float(elements[i + 5])
            sum_5th_column += float(elements[i + 6])
    
    # Update the values in the row
    elements[1] = str(sum_2nd_column)
    elements[3] = str(sum_3rd_column)
    elements[4] = str(sum_4th_column)
    elements[5] = str(sum_5th_column)
    
    # Construct the new row
    new_row = ','.join(elements)
    
    return new_row

# Function to update the entire CSV file
def update_csv(input_file, output_file):
    with open(input_file, 'r') as infile, open(output_file, 'w', newline='') as outfile:
        reader = csv.reader(infile)
        writer = csv.writer(outfile)
        
        # Iterate through each row in the input file
        for row in reader:
            # If the row is not empty, update it and write to the output file
            if row:
                updated_row = update_row(','.join(row))
                writer.writerow(updated_row.split(','))

# Example usage
update_csv('dish_metadata_cafe2.csv', 'output.csv')
```
